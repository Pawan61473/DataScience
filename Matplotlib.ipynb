```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_point = np.array([1,8])
y_point = np.array([3,10])
```

```python
plt.plot(x_point,y_point)
plt.show()
```

```python
plt.plot(x_point,y_point,'o')
plt.show()
```

```python
## If we dont provide x coordinate it automatically takes 
ypoints = np.array([3,5,1,2,10])
plt.plot(ypoints, 'o:r')
plt.show() 
```

```python
plt.plot(ypoints, linestyle = "dotted")
```

```python
x1 = np.array([1,5,8,10])
x2 = np.array([2,4,8,4])
y1 = np.array([7,1,4,0])
y2 = np.array([5,8,2,9])
```

```python
plt.plot(x1)
plt.plot(x2)
plt.plot(y1)
plt.plot(y2)
plt.title("My First Graph")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()
plt.show()
```

```python
## Subplot
x1 = np.array([1,5,8,5])
y1 = np.array([4,6,1,6])
plt.subplot(1,2,1)
plt.plot(x1,y1)
x2 = np.array([3,1,6,8])
y2 = np.array([4,2,5,7])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.suptitle('My Charts')
plt.show()
```

```python
# BarPlot
x=np.arange(0,10)
y=np.arange(11,21)
plt.bar(x,y)
```

```python
#Histogram
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
## Pie Chart
Labels = 'SQL','Python','Power BI','Machine Learning'
sizes = [210,245,231,200]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=Labels, colors=colors,autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
