```python
from bs4 import BeautifulSoup
import requests, openpyxl
```

```python
url = "https://www.imdb.com/list/ls009997493/"

source = requests.get(url)
```

```python
soup = BeautifulSoup(source.text, "html")
```

```python
movies = soup.find('div', class_="lister-list").find_all('div', class_='lister-item mode-detail')
len(movies)
```

```python
Ranks = []
Names = []
Years = []
Ratings = []
for movie in movies:
  rank = movie.find('div', class_= 'lister-item-content').h3.find('span', class_= "lister-item-index unbold text-primary").text
  Ranks.append(rank)
  name = movie.find('div', class_='lister-item-content').h3.a.text
  Names.append(name)
  year = movie.find('div', class_= 'lister-item-content').h3.find('span', class_= "lister-item-year text-muted unbold").text
  Years.append(year)
  rating = movie.find('div', class_= 'lister-item-content').find('div', class_= "ipl-rating-widget").find('div',class_="ipl-rating-star small").find('span', class_="ipl-rating-star__rating").text
  Ratings.append(rating)
```

```python
import pandas as pd
final_data = pd.DataFrame({"Rank":Ranks,"Name":Names,"Year":Years,"Rating":Ratings})
```
